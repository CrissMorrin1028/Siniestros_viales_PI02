Renombramos las columnas del dataframe **victimas** para que coincidan con las columnas del dataframe **hechos**

```python
# renombrar la columna ID_hecho' a ID
victimas.rename(columns={'ID_hecho':'ID'}, inplace=True)
```

```python
# Definir las columnas en las que queremos unir los DataFrames
columnas = ['ID', 'FECHA', 'AAAA', 'MM', 'DD', 'VICTIMA']

# Unir los DataFrames
datos_unidos = pd.merge(hechos, victimas, on=columnas, how='inner')
```

```python
datos_unidos.head(5)
```

```python
# ID unicos de la columna 'id' en el archivo 'hechos'
datos_unidos['ID'].nunique()
```

```python
# Unir los DataFrames con una unión externa
datos_unidos_externa = pd.merge(hechos, victimas, on=columnas, how='outer', indicator=True)

# Encontrar las filas que no se incluyeron en la unión
hechos_no_incluidos = datos_unidos_externa[datos_unidos_externa['_merge'] == 'left_only']
victimas_no_incluidas = datos_unidos_externa[datos_unidos_externa['_merge'] == 'right_only']
```

### **ETL** **PRINCIPAL**



```python
import pandas as pd

# Leer el archivo xlsx
xlsx_file = pd.ExcelFile('Datasets/homicidios.xlsx')

# Leer la página "Victimas" del archivo xlsx
df = pd.read_excel(xlsx_file, sheet_name='VICTIMAS')

# Convierte la columna 'FECHA_FALLECIMIENTO' a un objeto de fecha y hora
# Los valores que no se pueden convertir se reemplazan con NaT
df['FECHA_FALLECIMIENTO'] = pd.to_datetime(df['FECHA_FALLECIMIENTO'], errors='coerce')

# Formatea la columna 'FECHA_FALLECIMIENTO' como una cadena de fecha
# Los valores NaT se convierten a 'SD'
df['FECHA_FALLECIMIENTO'] = df['FECHA_FALLECIMIENTO'].dt.strftime('%Y-%m-%d').replace('NaT', 'SD')

# Guardar el dataframe en un archivo csv
df.to_csv('Homicidios/Victimas.csv', index=False)
```
